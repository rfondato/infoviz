```python
import pandas as pd
```

```python
data = pd.read_csv("./anatomia_desarrollo.csv")
```

```python
data
```

## Procesamiento de pivot por tareas:

```python
tareas_pivot = data[["Dia","Fecha", "Tarea", "Tiempo"]].pivot(index=["Dia", "Fecha"], columns="Tarea", values="Tiempo").reset_index()
```

```python
tareas_pivot
```

```python
tareas_pivot["Fecha"] = pd.to_datetime(tareas_pivot["Fecha"], format="%d-%m-%Y")
```

```python
tareas_pivot = tareas_pivot.sort_values(by=['Dia']).set_index('Dia')
```

```python
tareas_pivot.info()
```

```python
tareas_pivot
```

```python
tareas_pivot.to_csv("./tareas_pivot.csv")
```

## Procesamiento de pivot por categorías:

```python
cat_pivot = data[["Dia", "Fecha", "Categoría", "Tiempo"]].groupby(["Dia", "Fecha", "Categoría"]).sum().reset_index()
```

```python
cat_pivot
```

```python
cat_pivot = cat_pivot.pivot(index=["Dia", "Fecha"], columns="Categoría", values="Tiempo").reset_index()
cat_pivot
```

```python
cat_pivot["Fecha"] = pd.to_datetime(cat_pivot["Fecha"], format="%d-%m-%Y")
cat_pivot = cat_pivot.sort_values(by=['Dia']).set_index('Dia')
cat_pivot
```

```python
cat_pivot.info()
```

```python
cat_pivot.to_csv("./categorias_pivot.csv")
```
